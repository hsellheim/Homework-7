# Part Two

## Scrape the content of https://www.lemonde.fr/ and save it as a CSV. We want: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.

```python
from bs4 import BeautifulSoup
import requests

response = requests.get("https://www.lemonde.fr")
doc = BeautifulSoup(response.text)
```

```python
doc.find_all('p')
```

```python
items = doc.find_all(class_='article')

for item in items:
    print("------")
    print(item)
```

```python
rows = []

for item in items:
    row = {}
    row['headline'] = item.find(class_='article__title').text
    try:
        row['image'] = item.find('img').get('data-lazy', None)
    except:
        row['image'] = ('No image found')
    author_name = item.find(class_='article__author-name')
    if author_name:
        row['author name'] = author_name.text
    else: 
        row['author name'] = ('No author name found')
    row['url'] = item.select_one('a[href]')
    subheader = item.find(class_='article__desc article__subheader')
    if subheader:
        row['subheader'] = subheader.text
    else: 
        row['subheader'] = ('Now subheader found')
    premium = item.find(class_='sr-only')
    if premium:
        row['premium'] = ('yes')
    else:
        row['premium'] = ('no')
    article_type = item.find(class_='article__type')
    if article_type:
      row['arcticle type'] = article_type.text
    else: 
        row['arcticle type'] = ('No article type found')
    rows.append(row)
len(rows)
```

```python
import pandas as pd

df = pd.json_normalize(rows)
df.head()
```

```python
df.to_csv("lemonde.csv", index=False)
```
